```python
import os
import numpy as np
import torch
import time
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, Dataset
from transformers import ViTForImageClassification, ViTFeatureExtractor
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from tensorflow.keras import layers, models
```

```python
# Directory and file handling
image_folder = "./someData"
image_list = [img for img in os.listdir(image_folder) if img.endswith('.jpg')]

# Function to derive age and gender labels from filenames
def get_labels_from_filename(file_name):
    parts = file_name.split('_')
    if len(parts) < 2:
        print(f"File skipped, not enough parts: {file_name}")
        return None, None
    try:
        age = int(parts[0])
        gender = int(parts[1])
    except ValueError:
        print(f"Conversion error in file: {file_name}")
        return None, None
    return age, gender

# Creating a DataFrame with image metadata
metadata = []
for img_file in image_list:
    age, gender = get_labels_from_filename(img_file)
    if age is not None and gender is not None:
        metadata.append({'img_file': img_file, 'age': age, 'gender': gender})

df_images = pd.DataFrame(metadata)

# Split dataset into train, validation, and test sets
train_set, temp_set = train_test_split(df_images, test_size=0.3, random_state=42)
valid_set, test_set = train_test_split(temp_set, test_size=0.5, random_state=42)

# Function to load and preprocess images
def preprocess_image(file_name, img_size=(224, 224)):
    path = os.path.join(image_folder, file_name)
    img = Image.open(path).resize(img_size)
    return np.array(img) / 255.0

# Data preparation for model training
def prepare_dataset(df):
    imgs = np.array([preprocess_image(row['img_file']) for _, row in df.iterrows()])
    ages = np.array(df['age'])
    genders = np.array(df['gender'])
    return imgs, ages, genders

# Load training, validation, and testing datasets
train_imgs, train_ages, train_genders = prepare_dataset(train_set)
valid_imgs, valid_ages, valid_genders = prepare_dataset(valid_set)
test_imgs, test_ages, test_genders = prepare_dataset(test_set)

# Visualize distributions of age and gender
plt.figure(figsize=(14, 7))

# Age Distribution
plt.subplot(1, 2, 1)
plt.hist(df_images['age'], bins=range(0, 120, 5), color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# Gender Distribution
plt.subplot(1, 2, 2)
plt.hist(df_images['gender'], bins=[0, 1, 2], color='salmon', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0.5, 1.5], labels=['Male', 'Female'])  # Set x-axis labels to 'Male' and 'Female'
plt.show()

# Age statistics by gender
age_stats_by_gender = df_images.groupby('gender')['age'].agg(['mean', 'std', 'min', 'max'])

# Output statistics
print("Age Statistics Grouped by Gender:")
print(age_stats_by_gender)

# Define CNN model architecture
def create_cnn():
    model = models.Sequential([
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mae'])
    return model

# Train the CNN model
cnn_model = create_cnn()
start_cnn_time = time.time()
cnn_training_history = cnn_model.fit(train_imgs, train_ages, epochs=10, validation_data=(valid_imgs, valid_ages))
cnn_duration = time.time() - start_cnn_time
print(f'Training Time for CNN: {cnn_duration:.2f} seconds')

# Visualize CNN training and validation MAE
plt.figure(figsize=(14, 7))
plt.plot(cnn_training_history.history['mae'], label='Training MAE')
plt.plot(cnn_training_history.history['val_mae'], label='Validation MAE')
plt.title('CNN: Training and Validation Mean Absolute Error')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()

# Define custom dataset for PyTorch
class ImageDataset(Dataset):
    def __init__(self, imgs, ages, genders):
        self.imgs = torch.tensor(imgs, dtype=torch.float32).permute(0, 3, 1, 2)
        self.ages = torch.tensor(ages, dtype=torch.float32)
        self.genders = torch.tensor(genders, dtype=torch.long)
        
    def __len__(self):
        return len(self.imgs)
    
    def __getitem__(self, index):
        return self.imgs[index], self.genders[index]

# ViT model initialization
feature_ext = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')
vit_model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224-in21k', num_labels=2)

train_dataset = ImageDataset(train_imgs, train_ages, train_genders)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(vit_model.parameters(), lr=1e-5)

# Transformer training function
def train_vit_model(model, loader, criterion, optimizer, num_epochs=10):
    model.train()
    epoch_losses = []
    vit_start_time = time.time()
    for epoch in range(num_epochs):
        epoch_loss_sum = 0
        for images, labels in loader:
            optimizer.zero_grad()
            preds = model(images).logits

            # Ensure integer labels
            labels = labels.to(torch.long)

            # Check for NaNs
            if torch.isnan(preds).any() or torch.isnan(labels).any():
                print("Detected NaNs in outputs or labels")
                continue

            loss = criterion(preds, labels)

            if torch.isnan(loss).any():
                print("Detected NaNs in loss")
                continue

            loss.backward()
            optimizer.step()
            epoch_loss_sum += loss.item()
            
        average_loss = epoch_loss_sum / len(loader)
        epoch_losses.append(average_loss)
        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {average_loss:.4f}')
        
    total_vit_time = time.time() - vit_start_time
    print(f'ViT Model Training Time: {total_vit_time:.2f} seconds')
    return epoch_losses

losses = train_vit_model(vit_model, train_loader, criterion, optimizer)

# Visualize transformer training loss
plt.figure(figsize=(14, 7))
plt.plot(losses, label='Training Loss')
plt.title('Transformer: Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Transformer validation function
def validate_vit_model(model, loader):
    model.eval()
    validation_losses = []
    with torch.no_grad():
        for images, labels in loader:
            predictions = model(images).logits
            loss = criterion(predictions, labels)
            validation_losses.append(loss.item())
    mean_val_loss = np.mean(validation_losses)
    return mean_val_loss

validation_dataset = ImageDataset(valid_imgs, valid_ages, valid_genders)
valid_loader = DataLoader(validation_dataset, batch_size=32, shuffle=False)

validation_loss = validate_vit_model(vit_model, valid_loader)
print(f'ViT Model Validation Loss: {validation_loss:.4f}')

# Visualize transformer training and validation loss
plt.figure(figsize=(14, 7))
plt.plot(losses, label='Training Loss')
plt.axhline(y=validation_loss, color='red', linestyle='--', label='Validation Loss')
plt.title('Transformer: Training vs. Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Function to display confusion matrix
def plot_confusion_mtx(preds, actual_labels):
    conf_matrix = confusion_matrix(actual_labels, preds, labels=[0, 1])
    display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['Male', 'Female'])
    display.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix for Gender Classification')
    plt.show()

# Predictions and confusion matrix for both models
cnn_predictions = (cnn_model.predict(test_imgs).flatten() > 0.5).astype(int)
vit_predictions = [torch.argmax(vit_model(torch.tensor(img).unsqueeze(0).float().permute(0, 3, 1, 2)).logits, dim=1).item() for img in test_imgs]

plot_confusion_mtx(cnn_predictions, test_genders)
plot_confusion_mtx(vit_predictions, test_genders)

# Compare model performance
def evaluate_models(cnn_mod, vit_mod, test_imgs, test_ages, test_genders):
    cnn_preds = (cnn_mod.predict(test_imgs).flatten() > 0.5).astype(int)
    
    vit_mod.eval()
    vit_preds = [torch.argmax(vit_mod(torch.tensor(img).unsqueeze(0).float().permute(0, 3, 1, 2)).logits, dim=1).item() for img in test_imgs]
    
    cnn_mae = np.mean(np.abs(cnn_preds - test_ages))
    vit_acc = np.mean(np.array(vit_preds) == test_genders)
    
    print(f'CNN MAE for Age Prediction: {cnn_mae}')
    print(f'ViT Accuracy for Gender Classification: {vit_acc}')
    print("CNN Model Summary:")
    cnn_mod.summary()
    print("ViT Model Summary:")
    print(vit_mod)

evaluate_models(cnn_model, vit_model, test_imgs, test_ages, test_genders)

# Visualize true vs. predicted ages
def visualize_age_predictions(cnn_mod, test_imgs, test_ages):
    cnn_preds = cnn_mod.predict(test_imgs).flatten()
    plt.figure(figsize=(14, 7))
    plt.scatter(test_ages, cnn_preds, alpha=0.5)
    plt.xlabel('Actual Age')
    plt.ylabel('Predicted Age')
    plt.title('Age Prediction: Actual vs. Predicted')
    plt.show()

visualize_age_predictions(cnn_model, test_imgs, test_ages)

# Display sample images from the dataset
def display_sample_images(df, num_samples=5):
    genders = df['gender'].unique()
    plt.figure(figsize=(14, 7))
    for index, gender in enumerate(genders):
        plt.subplot(1, len(genders), index + 1)
        samples = df[df['gender'] == gender].sample(num_samples)
        for img_file in samples['img_file']:
            img = preprocess_image(img_file)
            plt.imshow(img)
            plt.title('Male' if gender == 0 else 'Female')
            plt.axis('off')
    plt.show()

display_sample_images(df_images)
```
